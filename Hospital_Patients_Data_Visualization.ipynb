```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("indian_liver_patient.csv")

ages = df["Age"].dropna()
```

### Histogram: Age Distribution of Patients

```python
plt.figure(figsize=(10, 6))
counts, bins, _ = plt.hist(ages, bins=10, edgecolor='black', alpha=0.7)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.75)
plt.show()

max_index = np.argmax(counts)
common_age_group = (bins[max_index], bins[max_index + 1])
print("Most common age group:", common_age_group)
```

### Bar Chart: Age Group Distribution

```python
age_groups = pd.cut(ages, bins=bins, right=False)
bar_data = age_groups.value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(bar_data.index.astype(str), bar_data.values, edgecolor='black', alpha=0.7)
plt.title("Age Group Distribution (Bar Chart)")
plt.xlabel("Age Groups")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

### Line Chart: Age Distribution Trend

```python
plt.figure(figsize=(10, 6))
bin_centers = (bins[:-1] + bins[1:]) / 2
if len(bin_centers) > len(counts):
    bin_centers = bin_centers[:len(counts)]

plt.plot(bin_centers, counts, marker='o', linestyle='-', color='skyblue')
plt.title("Age Distribution Trend (Line Chart)")
plt.xlabel("Age (Years)")
plt.ylabel("Count")
plt.grid(True, alpha=0.75)
plt.show()
```

### Scatter Plot: Individual Patient Ages

```python
plt.figure(figsize=(10, 6))
plt.scatter(x=np.arange(len(ages)), y=ages, alpha=0.6, s=10)
plt.title("Scatter Plot of Patient Ages")
plt.xlabel("Patient Index")
plt.ylabel("Age")
plt.grid(True, alpha=0.75)
plt.show()
```

### Pie Chart: Age Group Share

```python
plt.figure(figsize=(8, 8))
plt.pie(bar_data.values, labels=bar_data.index.astype(str), autopct="%1.1f%%", startangle=90, pctdistance=0.85)
plt.title("Age Group Share (Pie Chart)")
plt.axis('equal')
plt.show()
```
