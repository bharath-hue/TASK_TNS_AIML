**Function Without Parameters and Without Return Value**

```python
def display():
    print("Hello World")
display()
```

**Function With Parameters**

```python
def greet(name):
    print("Hello", name)

greet("Bharath")
greet("Kumar")
```

**Function With Return Value**

```python
def add(a,b):
    return a*b

result=add(3,12)
print(result)
```

**Function With Parameters and Return Value**

```python
def square(n):
    return n*n

result=square(6)
print(result)
```

**Lambda Function**

```python
multiply=lambda a, b: a * b
print(multiply(5,2))
```

**Classes in Python**

```python
class Person:
   name="Bharath"
a=Person()
print(a.name)
```

**Constructor Class in Python**

```python
class student:

  def __init__(self,name,age):
    self.name = name
    self.age = age
s1 = student("Bharath",21)

print("Name :",s1.name)
print("Age :",s1.age)
```

```python
class student:
  def __init__(self,name):
    self.name = name

  def display(self):
    print("Name of the student: ",self.name)

s1 = student("Bharath")
s1.display()
```

```python
class Family:
 def __init__(self,mother,father):
  self.mother=mother
  self.father=father

 def display(self):
  print("Mother:",self.mother)
  print("Father:",self.father)

f=Family("Sita","Ram")
f.display()
```

```python
class Student:
 def __init__(self,name,marks):
  self.name=name
  self.marks=marks

 def display(self):
  if self.marks>=40:
   print(self.name,"Pass")
  else:
   print(self.name,"Fail")

s=Student("Bharath",75)
s.display()
```

```python
class Employee:
 def __init__(self,name,salary):
  self.name=name
  self.salary=salary

 def display(self):
  if self.salary>50000:
   print(self.name,"High salary")
  else:
   print(self.name,"Low salary")

e=Employee("Priya",60000)
e.display()
```
