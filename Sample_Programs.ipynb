Reverse a String

```python
text="Bharath"
rev=""
for ch in text:
    rev=ch+rev
print("Reversed string:",rev)
```

Reverse a Number

```python
num=12345
rev=0
while num>0:
    digit=num%10
    rev=rev*10+digit
    num=num//10
print("Reversed number:",rev)
```

Fibonacci Series

```python
n=10
a=0
b=1
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Maximum Number

```python
a=3
b=7
c=6
if a>b and a>c:
    print(a)
elif b>c:
    print(b)
else:
    print(c)
```

Swap Two Numbers Without Using a Third Variable

```python
a=1
b=3
a = a + b
b = a - b
a = a - b
print("After swapping:")
print("a =",a)
print("b =",b)
```
