```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Line Plot**

```python
x=[1,2,3,4,5]
y=[10,20,40,60,80]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.legend("Values")
plt.xlabel("Nums")
plt.ylabel("Rate")
plt.show()
```

**Bar Graph**

```python
students=['A','B','C','D']
marks=[85,90,78,95]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

**Histogram**

```python
ages = [18,19,17,20,21,29,22,23,24,22,26,25]

plt.hist(ages, bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

**Scatter Plot**

```python
hours = [1,2,3,4,5,6]
marks = [20,30,50,60,75,90]

plt.scatter(hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

**Pie Chart**

```python
labels = ['Bus','Bike','Car','Auto']
values = [40,25,20,15]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Modes of Transport")
plt.show()
```
