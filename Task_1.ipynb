```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
uploaded=files.upload()
```

```python
data=pd.read_csv('Global YouTube Statistics.csv', encoding='latin1')
```

```python
data.head()
```

```python
top_channels=data.sort_values(by='subscribers', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Youtuber', y='subscribers', data=top_channels)
plt.title("Top 10 YouTube channels by subscribers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
highest_views=data.sort_values(by='video views', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Youtuber', y='video views', data=highest_views)
plt.title("YouTube channels with Highest Video Views")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
country_creators = data['Country'].value_counts().head(10)
plt.figure(figsize=(10,6))
plt.pie(country_creators, labels=country_creators.index, autopct='%1.1f%%', startangle=90)
plt.title("Countries with the highest number of YouTube creators")
plt.axis('equal')
plt.show()
```
